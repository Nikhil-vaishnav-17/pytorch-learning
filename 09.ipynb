```python
import torch
import torch.nn as nn

torch.manual_seed(42)
```

```python
# Data — simple binary classification
X = torch.rand(200, 4)
y = (X[:, 0] + X[:, 1] > 1.0).long()   # class 1 if first two features sum > 1
```

```python
X_train, X_val = X[:160], X[160:]
y_train, y_val = y[:160], y[160:]
```

```python
# Model, loss, optimizer
model     = nn.Sequential(nn.Linear(4, 8), nn.ReLU(), nn.Linear(8, 2))
loss_fn   = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
EPOCHS = 30

for epoch in range(EPOCHS):

    # ── Training ──
    model.train()
    for i in range(0, len(X_train), 32):
        xb = X_train[i:i+32]
        yb = y_train[i:i+32]
        loss = loss_fn(model(xb), yb)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    # ── Evaluation ──
    model.eval()
    with torch.no_grad():
        # Training metrics
        train_logits = model(X_train)
        train_loss   = loss_fn(train_logits, y_train).item()
        train_acc    = (train_logits.argmax(1) == y_train).float().mean().item()

        # Validation metrics
        val_logits   = model(X_val)
        val_loss     = loss_fn(val_logits, y_val).item()
        val_acc      = (val_logits.argmax(1) == y_val).float().mean().item()

    if (epoch + 1) % 5 == 0:
        print(f"Epoch {epoch+1:2d} | "
              f"Train loss {train_loss:.3f}, acc {train_acc:.0%} | "
              f"Val loss {val_loss:.3f}, acc {val_acc:.0%}")
```
