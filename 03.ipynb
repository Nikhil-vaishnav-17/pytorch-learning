```python
import torch
```

```python
# A normal tensor — no tracking
x = torch.tensor([2.0])
print(x)

# A tensor PyTorch will track
x = torch.tensor([2.0], requires_grad=True)
print(x)
```

#### Think of 'w' as a single weight in a neural network <br> requires_grad=True tells PyTorch: "track everything done to this"


```python
w = torch.tensor([3.0], requires_grad=True)
w
```

#### A simple computation — like a loss function <br> loss = w^2  →  at w=3, loss = 9

```python
loss = w ** 2
print(loss)   # tensor([9.], grad_fn=<PowBackward0>)
#                                   ↑
#              Notice: PyTorch already recorded the operation!
```

#### Now ask PyTorch: "what is d(loss)/d(w)?"

```python
loss.backward()
```

#### The gradient is stored in w.grad

```python
print(w.grad)   # tensor([6.])
```

# More Realistic Example
#### Two weights — like two parameters in a small model

```python
w1 = torch.tensor([2.0], requires_grad=True)
w2 = torch.tensor([3.0], requires_grad=True)
```

#### A slightly more complex loss <br> loss = w1^2 + w2^3

```python
loss = w1 ** 2 + w2 ** 3
```

```python
loss.backward()
```

#### Gradients computed simultaneously for both

```python
print(w1.grad)   # tensor([4.])  ← d/dw1 of w1² = 2*w1 = 4
print(w2.grad)   # tensor([27.]) ← d/dw2 of w2³ = 3*w2² = 27
```

#### PyTorch accumulates gradients by default. So if you call .backward() twice, the gradients add up — which is almost never what you want:

```python
w = torch.tensor([3.0], requires_grad=True)

# First backward
loss = w ** 2
loss.backward()
print(w.grad)   # tensor([6.])  ✅ correct
```

```python
# Second backward WITHOUT zeroing first
loss = w ** 2
loss.backward()
print(w.grad)   # tensor([12.]) ❌ wrong! It added 6 to the previous 6
```

```python
# Correct way: zero out before each backward
w.grad.zero_()  # the underscore means "in-place operation"
loss = w ** 2
loss.backward()
print(w.grad)   # tensor([6.])  ✅ correct again
```
