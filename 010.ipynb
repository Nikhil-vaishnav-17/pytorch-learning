```python
import torch
import torch.nn as nn

torch.manual_seed(42)
```

```python
X = torch.randn(10, 4)       # 10 samples, 4 features
y = torch.randint(0, 2, (10,))  # Binary labels: 0 or 1
```

```python
# Train/val split: 8 train, 2 val
X_train, y_train = X[:8], y[:8]
X_val,   y_val   = X[8:], y[8:]
```

```python
# ── 2. Oversized model for 10 samples (will overfit easily) ──────────
class BigModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(4, 128),   # Way too many neurons for 10 samples
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU(),
            nn.Linear(128, 2)    # 2 output classes
        )

    def forward(self, x):
        return self.net(x)
```

```python
model = BigModel()
```

```python
# ── 3. Loss and optimizer ─────────────────────────────────────────────
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
# ── 4. Training loop — watch both losses ─────────────────────────────
for epoch in range(200):
    # --- Train ---
    model.train()
    optimizer.zero_grad()
    out = model(X_train)
    train_loss = criterion(out, y_train)
    train_loss.backward()
    optimizer.step()

    # --- Validate ---
    model.eval()
    with torch.no_grad():
        val_out = model(X_val)
        val_loss = criterion(val_out, y_val)

    # Print every 40 epochs
    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1:3d} | Train Loss: {train_loss.item():.8f} | Val Loss: {val_loss.item():.4f}")
```
