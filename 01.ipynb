```python
import torch
```

## --- Creating tensors ---

## A scalar (0D)

```python
a = torch.tensor(3.14)
print(a)          # tensor(3.1400)
print(a.shape)    # torch.Size([]) — empty shape = 0 dimensions
```

## A vector (1D) — just like a Python list

```python
b = torch.tensor([1.0, 2.0, 3.0])
print(b)          # tensor([1., 2., 3.])
print(b.shape)    # torch.Size([3]) — 3 elements
```

## A matrix (2D) — just like a 2D NumPy array

```python
c = torch.tensor([[1.0, 2.0],
                  [3.0, 4.0]])
print(c.shape)    # torch.Size([2, 2]) — 2 rows, 2 columns
```

# --- Useful creation shortcuts ---

```python
zeros = torch.zeros(3, 4)    # 3×4 matrix of all zeros
ones = torch.ones(2, 2)     # 2×2 matrix of all ones
rand = torch.rand(3, 3)     # 3×3 matrix of random values between 0 and 1
print(rand)
```

# --- Key property: dtype ---
## Tensors have a data type, just like NumPy dtypes

```python
x = torch.tensor([1.0, 2.0])
print(x.dtype)    # torch.float32 — the default for decimals
```

```python
y = torch.tensor([1, 2])
print(y.dtype)    # torch.int64 — integers default to int64
```
