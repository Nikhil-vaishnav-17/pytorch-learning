## Code: A minimal CNN in PyTorch

```python
import torch
import torch.nn as nn
```

```python
class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()
        
        # Conv layer: 1 input channel (grayscale), 16 filters, each 3x3
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, padding=1)

        # Pooling: takes 2x2 blocks and keeps the max value
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)

        # After conv+pool on a 28x28 image -> output is 16 channels of 14x14
        # Flatten: 16 * 14 * 14 = 3136
        self.fc1 = nn.Linear(16 * 14 * 14, 10) # 10 Output classes

    def forward(self, x):
        x = self.conv1(x)       # Apply convolution → shape: (batch, 16, 28, 28)
        x = torch.relu(x)       # Activation
        x = self.pool(x)        # Downsample → shape: (batch, 16, 14, 14)
        x = x.view(x.size(0), -1)  # Flatten → shape: (batch, 3136)
        x = self.fc1(x)         # Final classifier → shape: (batch, 10)
        return x
```

```python
model = SimpleCNN()
print(model)
```

```python
# Test with a fake image batch: 4 images, 1 channel, 28x28 pixels
dummy = torch.randn(4, 1, 28, 28)
out = model(dummy)
print("Output shape:", out.shape)  # Should be (4, 10)
```
